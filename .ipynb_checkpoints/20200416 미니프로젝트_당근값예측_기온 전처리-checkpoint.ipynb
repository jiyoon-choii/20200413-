```python
carrot_weather = pd.read_csv('./Data/0416_mini_project/carrot_weather.csv')
```

# Data : carrot_weather

## 원하는 날짜만큼 불러오기 (2019-06-24 ~ 2020-04-15)

```python
carrot_weather1 = carrot_weather[carrot_weather['날짜'].values > '2019-06-24']
carrot_weather1.head()
```

## 가격 결측치 채우기('bfill')

```python
carrot_weather1['가격'].fillna(method='bfill', inplace=True)
carrot_weather1.head(15)
```

## 날짜 외 컬럼을 float 혹은 int 형식으로 바꾸기 

```python
carrot_weather1.dtypes
```

```python
carrot_weather1['날짜'] = pd.to_datetime(carrot_weather1['날짜'])
carrot_weather1.head()
```

```python
def change_float(x) : 
    try :
        tmp = float(x)
    except : 
        if len(x) > 1 :
            tmp = float(x[1:]) * -1
        else :
            tmp = None
    return tmp 
```

```python
carrot_weather1['현재기온'] = carrot_weather1['현재기온'].apply(change_float)
carrot_weather1['이슬점온도'] = carrot_weather1['이슬점온도'].apply(change_float)
carrot_weather1['체감온도'] = carrot_weather1['체감온도'].apply(change_float)
carrot_weather1['습도'] = carrot_weather1['습도'].apply(change_float)
carrot_weather1['풍속'] = carrot_weather1['풍속'].apply(change_float)
carrot_weather1['해면기압'] = carrot_weather1['해면기압'].apply(change_float)
```
