### 데이터프레임의 열단위 수치연산 및 데이터 타입 다루기 
- 클릭수(반올림, 일의 자리로 변경) 
- 클릭률
- 평균 클릭 비용(VAT 포함, 원) 

```python
# 클릭수 : 한 번의 클릭이 총 몇 번 이루어졌는지 나타내는 지표 => 반올림 
clk = round(df['클릭수'], 0)
```

```python
clk.astype(int)
```

```python
df.head()
```

```python
# CTR 
df['클릭률(%)'] = df['클릭수'] / df['노출수'] * 100 
```

```python
# CPC 
cpc = round(df['평균클릭비용(VAT포함,원)'], 0)
df['총비용(vat포함, 원)'] = cpc / df['클릭률(%)'] * 100
```

```python
pd.set_option('display.float_format', '{:.2f}'.format) # 형식 float 형식으로 
```

```python
df.describe()
```

## 중점관리키워드 추출 
- 노출수, 클릭 수 기준 상위 5%에 해당하는 키워드 추출 
    - 95 백분위 수 찾기 (quantile 함수 사용)
    - 상위 5%에 해당하는 키워드 추출 

```python
imp = df['노출수']
```

```python
# 95 백분위 수 찾기 
imp.quantile()
```

```python
imp.median()
```

```python
print(imp.quantile(0))
print(imp.min())
```

```python
print(imp.quantile(1))
print(imp.max())
```

```python
imp.quantile(0.95)
```

```python
imp = imp[imp >= imp.quantile(0.95)]
imp
```

```python
df_index = df.set_index('키워드')
```

```python
df_index
```

```python
imp = df_index['노출수']
imp = imp[imp >=  imp.quantile(0.95)]
```

```python
imp.head()
```

```python
clk = df_index['클릭수']
clk = clk[clk >= clk.quantile(0.95)]
```

```python
clk.head()
```

```python
# 노출수와 클릭수 모두 95 백분위 수 이상인 키워드 
imp = df_index['노출수']
clk = df_index['클릭수']
```

```python
result = df_index[(imp>=imp.quantile(0.95)) & (clk >= clk.quantile(0.95))]
```

```python
result.head()
```

```python
result.index
```

## 노출수, 클릭수 기준 상위 5%의 키워드가 아닌데도 비용이 많이 쓰이는 키워드 
- 노출수 95 백분위수 미만
- 클릭수 95 백분위수 미만
- 총비용 85 백분위수 이상
- 총비용 95 백분위수 미만 

```python
cost = df_index['총비용(VAT포함,원)']
```

```python
result = df_index[(imp < imp.quantile(0.95)) & (clk < clk.quantile(0.95)) & (cost >= cost.quantile(0.85))]
result
```

```python
result.index
```

### 광고그룹 분석 
- as-is : 키워드 기준 분석 
    - 키워드의 노출 수 / 클릭 수 / 총 비용 .... 
- to-bo : 광고그룹 기준 분석 
    - 데이터의 구성을 광고그룹 기준으로 변경 
    - 광고그룹의 노출 수/ 클릭 수 / 총 비용...
    - groupby 함수 사용 

```python
grouped = df.groupby('광고그룹')
```

```python
grouped.count()
```

```python
grouped.mean()
```

```python
grouped.median()
```
